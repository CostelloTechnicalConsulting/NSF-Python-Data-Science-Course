## Get the data

```python
import pandas as pd
awd_url = 'https://onlcfiles.blob.core.windows.net/clientfiles/NSF/nsf-awards-2010-2025.csv'
awards = pd.read_csv(awd_url)
```

```python
awards.head()
```

```python
awards.sort_values('awd_titl_txt',inplace=True)
awards.head(10)
```

## Use tab completion and tooltips

```python
awards.sort_values('po_email', ascending=False).head()
```

## Clean the data

### Keep only the rows with "Grant" in the tran_type column

```python
awards = awards.query('tran_type=="Grant"')
awards.head()
```

### Rename some columns

```python
awards.rename(columns={'agcy_id':'Agency_ID',
                      'awd_titl_txt':'Title'},inplace=True)
awards.head()
```

### Convert date columns to datetime objects

```python
date_cols = ['awd_eff_date', 'awd_exp_date','awd_min_amd_letter_date', 'awd_max_amd_letter_date']
awards[date_cols] = awards[date_cols].apply(pd.to_datetime)
awards.head(3)
```

### Try two plots

```python
awards.plot.line(x='awd_eff_date',y=['tot_intn_awd_amt','awd_amount'])
```

```python
%%time
# Note use of backticks (`) to escape the column name
awards.query('`inst.inst_state_code` == "TX"') \
    .plot(x='awd_eff_date',y=['tot_intn_awd_amt','awd_amount'])
```
